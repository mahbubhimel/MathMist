```python
import json
import os
import re
from datetime import datetime
from tqdm import tqdm

# Ollama client
try:
    import ollama
except Exception as e:
    raise ImportError("The 'ollama' package is required. Install it and make sure the Ollama daemon is running.") from e

# -------------------------
# Enhanced Extractor for Decision (YES/NO)
# -------------------------
class SolutionVerificationExtractor:
    """
    Robust extractor for solution verification outputs.
    Extracts:
    - Decision (YES or NO) from multiple possible formats
    - Full justification text (excluding decision markers)
    """

    # Comprehensive patterns for YES/NO detection
    YES_PATTERNS = [
        # Standard tags
        r'<final>\s*YES\s*</final>',
        r'<final>\s*yes\s*</final>',
        r'<final>\s*1\s*</final>',

        # Markdown bold
        r'\*\*YES\*\*',
        r'\*\*Yes\*\*',
        r'\*\*yes\*\*',
        r'\*\*INCORRECT\*\*',
        r'\*\*Incorrect\*\*',
        r'\*\*ERROR\*\*',
        r'\*\*Error\*\*',

        # Dollar signs
        r'\$\$YES\$\$',
        r'\$\$yes\$\$',
        r'\$YES\$',
        r'\$yes\$',

        # Multiple asterisks
        r'\*\*\*\*YES\*\*\*\*',
        r'\*\*\*\*yes\*\*\*\*',

        # Decision labels
        r'(?:DECISION|Decision|ANSWER|Answer|FINAL|Final|VERDICT|Verdict|RESULT|Result)[\s:]+YES',
        r'(?:DECISION|Decision|ANSWER|Answer|FINAL|Final|VERDICT|Verdict|RESULT|Result)[\s:]+yes',

        # With brackets/parentheses
        r'\[YES\]',
        r'\[yes\]',
        r'\(YES\)',
        r'\(yes\)',

        # Standalone on line
        r'^YES$',
        r'^yes$',
    ]

    NO_PATTERNS = [
        # Standard tags
        r'<final>\s*NO\s*</final>',
        r'<final>\s*no\s*</final>',
        r'<final>\s*0\s*</final>',

        # Markdown bold
        r'\*\*NO\*\*',
        r'\*\*No\*\*',
        r'\*\*no\*\*',
        r'\*\*CORRECT\*\*',
        r'\*\*Correct\*\*',
        r'\*\*NO ERROR\*\*',
        r'\*\*No Error\*\*',

        # Dollar signs
        r'\$\$NO\$\$',
        r'\$\$no\$\$',
        r'\$NO\$',
        r'\$no\$',

        # Multiple asterisks
        r'\*\*\*\*NO\*\*\*\*',
        r'\*\*\*\*no\*\*\*\*',

        # Decision labels
        r'(?:DECISION|Decision|ANSWER|Answer|FINAL|Final|VERDICT|Verdict|RESULT|Result)[\s:]+NO',
        r'(?:DECISION|Decision|ANSWER|Answer|FINAL|Final|VERDICT|Verdict|RESULT|Result)[\s:]+no',

        # With brackets/parentheses
        r'\[NO\]',
        r'\[no\]',
        r'\(NO\)',
        r'\(no\)',

        # Standalone on line
        r'^NO$',
        r'^no$',
    ]

    @staticmethod
    def extract_decision(text: str) -> str:
        """
        Extract decision (YES or NO) using comprehensive pattern matching.
        Returns 'YES', 'NO', or empty string if not found.
        Priority: Check YES patterns first, then NO patterns.
        """
        if not text:
            return ""

        # Check for YES patterns
        for pattern in SolutionVerificationExtractor.YES_PATTERNS:
            if re.search(pattern, text, re.MULTILINE | re.IGNORECASE):
                return "YES"

        # Check for NO patterns
        for pattern in SolutionVerificationExtractor.NO_PATTERNS:
            if re.search(pattern, text, re.MULTILINE | re.IGNORECASE):
                return "NO"

        # Fallback: search in last few lines
        return SolutionVerificationExtractor.fallback_extract_decision(text)

    @staticmethod
    def fallback_extract_decision(text: str) -> str:
        """
        Fallback: try to find YES or NO in the last few lines.
        """
        if not text:
            return ""

        lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

        # Check last 5 lines for YES/NO indicators
        for line in reversed(lines[-5:]):
            # Direct YES/NO match
            if re.search(r'\bYES\b', line, re.IGNORECASE):
                return "YES"
            if re.search(r'\bNO\b', line, re.IGNORECASE):
                return "NO"

            # Check for error/correct indicators
            if re.search(r'\b(incorrect|error|wrong|mistake)\b', line, re.IGNORECASE):
                return "YES"
            if re.search(r'\b(correct|accurate|valid|right)\b', line, re.IGNORECASE):
                return "NO"

        return ""

    @staticmethod
    def extract_justification(text: str, decision: str) -> str:
        """
        Extract the full justification text, removing decision markers.
        Removes all pattern matches that indicate YES/NO decision.
        """
        if not text:
            return ""

        justification = text

        # Remove <think>...</think> tags if present
        justification = re.sub(r'<think>.*?</think>', '', justification, flags=re.DOTALL | re.IGNORECASE)

        # Remove all decision markers
        all_patterns = SolutionVerificationExtractor.YES_PATTERNS + SolutionVerificationExtractor.NO_PATTERNS

        for pattern in all_patterns:
            justification = re.sub(pattern, '', justification, flags=re.MULTILINE | re.IGNORECASE)

        # Remove common decision section headers
        justification = re.sub(
            r'(?:DECISION|Decision|ANSWER|Answer|FINAL|Final|VERDICT|Verdict|RESULT|Result)[\s:]+(?:YES|NO|yes|no)',
            '',
            justification,
            flags=re.IGNORECASE
        )

        # Clean up extra whitespace and newlines
        justification = re.sub(r'\n\s*\n\s*\n+', '\n\n', justification)  # Remove excessive newlines
        justification = re.sub(r'[ \t]+', ' ', justification)  # Normalize spaces
        justification = justification.strip()

        return justification


# -------------------------
# Ollama-based Solution Verifier (Enhanced Prompt)
# -------------------------
class OllamaSolutionVerifier:
    """
    Solution verifier that uses Ollama with prompted chain-of-thought reasoning.
    - Evaluates if a proposed solution is correct or contains errors IN THE STEPS
    - Outputs binary decision: NO (No error) or YES (Error found)
    - Provides justification for the decision
    """

    def __init__(self, model_name="gpt-oss:20b"):
        """
        model_name: the Ollama model reference (e.g., "gpt-oss:20b")
        """
        self.model_name = model_name
        self.client = self._load_client()

    def _load_client(self):
        print(f"Initializing Ollama client for model: {self.model_name}")
        client = ollama.Client()
        return client

    def cleanup(self):
        if hasattr(self, 'client'):
            del self.client

    def _create_verification_prompt(self, question: str, proposed_solution: str, exact_answer: str) -> str:
        """
        Create an enhanced verification prompt that EXPLICITLY checks solution steps.
        Forces YES/NO responses with detailed justifications about ERRORS IN REASONING.
        Does NOT compare final answers to avoid flagging typos as errors.
        """
        prompt = f"""🎯 **CRITICAL VERIFICATION TASK** 🎯

You are an expert mathematician tasked with verifying a mathematical solution. This is EXTREMELY IMPORTANT!

**YOUR MISSION:**
Verify if the proposed solution contains ANY ERRORS in its reasoning, calculations, or methodology.

⚠️ **CRITICAL INSTRUCTIONS** ⚠️
1. **CHECK EVERY STEP FOR MATHEMATICAL AND LOGICAL CORRECTNESS**
2. **DO NOT FLAG TYPOS OR FORMATTING ERRORS AS MISTAKES**
3. **IGNORE THE FINAL ANSWER - ONLY VERIFY THE SOLUTION STEPS**

Focus ONLY on:
✓ Mathematical calculations (are they computed correctly?)
✓ Formula applications (are formulas used correctly?)
✓ Logical reasoning (does each step follow logically?)
✓ Conceptual understanding (are concepts applied correctly?)

DO NOT consider as errors:
✗ Typos in the final answer
✗ Formatting issues
✗ Minor notation differences
✗ Answer presentation style

A solution has errors ONLY if:
- Calculation mistakes in the steps → ERROR (say <final>YES</final>)
- Incorrect formulas applied → ERROR (say <final>YES</final>)
- Logical fallacies in reasoning → ERROR (say <final>YES</final>)
- Wrong mathematical operations → ERROR (say <final>YES</final>)
- Conceptual misunderstandings → ERROR (say <final>YES</final>)

**⚠️ MANDATORY OUTPUT FORMAT ⚠️**

You MUST respond with EXACTLY ONE of these:
- **<final>YES</final>** = The solution contains errors (incorrect calculations, wrong logic, invalid steps, mathematical mistakes)
- **<final>NO</final>** = The solution is completely correct (all steps valid, correct reasoning, proper methodology)

**📋 RESPONSE STRUCTURE (STRICTLY FOLLOW THIS):**

1️⃣ **Step-by-Step Analysis of EACH STEP:**
[Go through EVERY step of the proposed solution. Check:]
   - Is the mathematical operation correct?
   - Is the formula used correctly?
   - Are calculations accurate?
   - Is the logic sound?
   - Are there any conceptual errors?
   - Mark any errors you find with ❌

2️⃣ **Error Detection:**
[List ALL errors found in the STEPS (not the final answer)]
   - Calculation errors: [describe any]
   - Logical errors: [describe any]
   - Methodological errors: [describe any]
   - Conceptual errors: [describe any]

3️⃣ **Your Decision (CRITICAL - USE ONE OF THESE FORMATS):**
FIRST LINE (decision token ONLY — nothing else on this line)
- The first line of your entire response must be exactly one of these two literal tokens:
  <final>YES</final>
  OR
  <final>NO</final>
- Requirements for the first line:
  - It must contain only the chosen token (no leading/trailing spaces, no punctuation, no additional characters).
  - It must be the very first line of the reply (no blank lines, no comments, nothing before it).
  - Use UPPERCASE inside the tags exactly as shown.

4️⃣ **Detailed Justification:**
- If **<final>YES</final>**: Explain WHICH steps are wrong, WHAT the errors are, and WHY they are incorrect
- If **<final>NO</final>**: Explain WHY the solution steps are correct, confirming all key steps are mathematically valid

---

**📝 MATH PROBLEM:**
{question}

**🔍 PROPOSED SOLUTION TO VERIFY:**
{proposed_solution}

---

🚨 **CRITICAL REMINDERS:** 🚨
✓ Check EVERY SINGLE STEP in the solution process
✓ Focus on MATHEMATICAL and LOGICAL correctness only
✓ IGNORE typos, formatting, or answer presentation issues
✓ 📌 You MUST output <final>YES</final> or <final>NO</final> as your decision in the first line.
✓ If errors exist, explain WHAT is wrong and WHERE in the steps
✓ If correct, explain WHY all steps are mathematically valid

Begin your verification now:
"""
        return prompt

    def _generate_verification(self, prompt_text: str) -> str:
        """
        Call ollama.Client.chat to generate verification output.
        Uses think=False but the prompt forces chain-of-thought.
        """
        messages = [{"role": "user", "content": prompt_text}]

        # Call Ollama client.chat with think=False
        try:
            resp = self.client.chat(model=self.model_name, messages=messages, think=False)
        except TypeError:
            try:
                resp = self.client.chat(self.model_name, messages=messages, think=False)
            except Exception:
                resp = self.client.chat(self.model_name, messages)

        # Normalize response into a string
        full_output = ""
        if isinstance(resp, dict):
            if 'message' in resp and isinstance(resp['message'], dict) and 'content' in resp['message']:
                full_output = resp['message']['content']
            elif 'choices' in resp and isinstance(resp['choices'], (list, tuple)) and resp['choices']:
                choice = resp['choices'][0]
                if isinstance(choice, dict) and 'message' in choice and isinstance(choice['message'], dict):
                    full_output = choice['message'].get('content', '')
                else:
                    full_output = str(choice)
            else:
                full_output = str(resp)
        else:
            try:
                full_output = resp.message.content
            except Exception:
                full_output = str(resp)

        if isinstance(full_output, bytes):
            full_output = full_output.decode('utf-8', errors='ignore')
        return (full_output or "").strip()

    def verify_solution(self, question: str, proposed_solution: str, exact_answer: str):
        """
        Verify a proposed solution and return decision + justification.

        Returns dict with:
        - raw_output: complete model response (for internal use)
        - decision: 'YES' or 'NO' (or empty if extraction failed)
        - llm_generated_error_desc: full justification if decision is 'YES', else empty
        - llm_no_error_justification: full justification if decision is 'NO', else empty
        - extracted_answer: the answer extracted from proposed solution
        """
        prompt = self._create_verification_prompt(question, proposed_solution, exact_answer)
        raw_output = self._generate_verification(prompt)

        # Extract decision
        decision = SolutionVerificationExtractor.extract_decision(raw_output)

        # Extract justification (full text minus decision markers)
        justification = SolutionVerificationExtractor.extract_justification(raw_output, decision)

        # Assign to appropriate field based on decision
        llm_generated_error_desc = ""
        llm_no_error_justification = ""

        if decision == 'YES':
            llm_generated_error_desc = justification
        elif decision == 'NO':
            llm_no_error_justification = justification

        return {
            "raw_output": raw_output,  # Keep for failed extractions
            "decision": decision,
            "llm_generated_error_desc": llm_generated_error_desc,
            "llm_no_error_justification": llm_no_error_justification
        }


# -------------------------
# Dataset Processor for Solution Verification
# -------------------------
class SolutionVerificationProcessor:
    def __init__(self, verifier: OllamaSolutionVerifier, failed_folder=None):
        self.verifier = verifier
        self.failed_folder = failed_folder or "failed_verifications"
        os.makedirs(self.failed_folder, exist_ok=True)

    def process_dataset(self, dataset_path, output_base_path, start_idx=0, end_idx=None,
                        folder_name=None, create_timestamped_folder=True):
        dataset = self._load_dataset(dataset_path)
        if end_idx is None:
            end_idx = len(dataset)

        output_folder = self._create_output_folder(output_base_path, folder_name, start_idx, end_idx, create_timestamped_folder)
        results = []

        print(f"Processing verification for problems {start_idx} to {end_idx-1} ({end_idx-start_idx} total)")
        print(f"Output will be saved in: {output_folder}")

        for idx in tqdm(range(start_idx, min(end_idx, len(dataset)))):
            problem = dataset[idx]
            try:
                result_entry = self._process_single_problem(idx, problem)
                results.append(result_entry)
                self._print_progress(idx, result_entry)
                if (idx - start_idx + 1) % 10 == 0:
                    self._save_intermediate_results(results, output_folder, idx - start_idx + 1)
            except Exception as e:
                print(f"Error processing problem {idx+1}: {str(e)}")
                error_entry = self._create_error_entry(idx, problem, str(e))
                results.append(error_entry)

        final_output_path = self._save_final_results(results, output_folder, start_idx, end_idx)
        self._create_summary_file(results, output_folder, dataset_path, start_idx, end_idx)
        return results, output_folder

    def _create_output_folder(self, base_path, folder_name, start_idx, end_idx, add_timestamp):
        if folder_name is None:
            folder_name = f"verification_results_{start_idx}_to_{end_idx-1}"
        if add_timestamp:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            folder_name = f"{folder_name}_{timestamp}"
        output_folder = os.path.join(base_path, folder_name)
        os.makedirs(output_folder, exist_ok=True)
        return output_folder

    def _load_dataset(self, dataset_path):
        dataset = []
        with open(dataset_path, 'r', encoding='utf-8') as f:
            for line in f:
                if line.strip():
                    dataset.append(json.loads(line))
        return dataset

    def _process_single_problem(self, idx, problem):
        question = problem.get("Question", "")
        exact_answer = problem.get("Exact Answer", "")
        wrong_solution = problem.get("Wrong_Solution_Version_3", "")

        if not wrong_solution:
            print(f"\nWarning: Problem {idx+1} has no 'Wrong_Solution_Version_3' field!")
            wrong_solution = ""

        print(f"\nProcessing Problem {idx+1} Verification")

        # Verify the solution
        verification_result = self.verifier.verify_solution(question, wrong_solution, exact_answer)

        raw_output = verification_result.get('raw_output', '')
        decision = verification_result.get('decision', '')
        error_desc = verification_result.get('llm_generated_error_desc', '')
        no_error_justification = verification_result.get('llm_no_error_justification', '')

        # Save failed extraction if decision wasn't extracted
        if not decision:
            fname = f"failed_verification_{idx}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
            fpath = os.path.join(self.failed_folder, fname)
            with open(fpath, 'w', encoding='utf-8') as f:
                json.dump({
                    "index": idx,
                    "question": question,
                    "wrong_solution": wrong_solution,
                    "exact_answer": exact_answer,
                    "raw_output": raw_output,
                    "decision": decision,
                    "llm_generated_error_desc": error_desc,
                    "llm_no_error_justification": no_error_justification
                }, f, ensure_ascii=False, indent=2)
            print(f"WARNING: Could not extract decision. Saved to {fpath}")

        # Create result entry with ALL original fields plus new ones
        result_entry = problem.copy()  # Keep all original fields

        # Add new fields
        result_entry["Decision"] = decision
        result_entry["llm_generated_error_desc"] = error_desc
        result_entry["llm_no_error_justification"] = no_error_justification

        return result_entry

    def _create_error_entry(self, idx, problem, error_msg):
        result_entry = problem.copy()
        result_entry["Decision"] = ""
        result_entry["llm_generated_error_desc"] = ""
        result_entry["llm_no_error_justification"] = ""
        return result_entry

    def _print_progress(self, idx, result_entry):
        decision = result_entry.get('Decision', '')
        error_desc = result_entry.get('llm_generated_error_desc', '')
        no_error_just = result_entry.get('llm_no_error_justification', '')

        decision_label = "ERROR FOUND" if decision == 'YES' else "CORRECT" if decision == 'NO' else "UNKNOWN"
        print(f"Decision: {decision} ({decision_label})")
        if error_desc:
            preview = error_desc[:150] + "..." if len(error_desc) > 150 else error_desc
            print(f"Error Description: {preview}")
        if no_error_just:
            preview = no_error_just[:150] + "..." if len(no_error_just) > 150 else no_error_just
            print(f"Correct Justification: {preview}")

    def _save_intermediate_results(self, results, output_folder, count):
        temp_filename = f'intermediate_verifications_{count}.json'
        temp_output_path = os.path.join(output_folder, temp_filename)
        with open(temp_output_path, 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=2)
        print(f"Saved intermediate results to {temp_output_path}")

    def _save_final_results(self, results, output_folder, start_idx, end_idx):
        final_filename = f'final_verifications_{start_idx}_to_{end_idx-1}.json'
        final_output_path = os.path.join(output_folder, final_filename)
        with open(final_output_path, 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=2)
        print(f"\nVerification complete. Results saved to {final_output_path}")
        print(f"Total problems processed: {len(results)}")
        return final_output_path

    def _create_summary_file(self, results, output_folder, dataset_path, start_idx, end_idx):
        decisions = [r.get('Decision', '') for r in results]
        errors_found = decisions.count('YES')
        no_errors = decisions.count('NO')
        failed_extractions = len([d for d in decisions if d not in ['YES', 'NO']])

        summary_data = {
            "verification_info": {
                "dataset_path": dataset_path,
                "start_index": start_idx,
                "end_index": end_idx - 1,
                "total_processed": len(results),
                "processing_timestamp": datetime.now().isoformat(),
                "output_folder": output_folder
            },
            "statistics": {
                "errors_found": errors_found,
                "correct_solutions": no_errors,
                "failed_extractions": failed_extractions,
                "error_rate": f"{(errors_found/len(results)*100):.1f}%" if results else "0%",
                "correct_rate": f"{(no_errors/len(results)*100):.1f}%" if results else "0%",
                "extraction_success_rate": f"{((errors_found+no_errors)/len(results)*100):.1f}%" if results else "0%"
            }
        }
        summary_path = os.path.join(output_folder, 'verification_summary.json')
        with open(summary_path, 'w', encoding='utf-8') as f:
            json.dump(summary_data, f, ensure_ascii=False, indent=2)
        print(f"Verification summary saved to {summary_path}")
        print(f"Errors found: {errors_found}/{len(results)} ({summary_data['statistics']['error_rate']})")
        print(f"Correct solutions: {no_errors}/{len(results)} ({summary_data['statistics']['correct_rate']})")
        print(f"Extraction success: {summary_data['statistics']['extraction_success_rate']}")


# -------------------------
# Main (example usage)
# -------------------------
def main():
    # Update dataset_path and output_base_path to match your environment
    dataset_path = "/kaggle/input/nctb-perturbation-data/English_Wrong_3_150.jsonl"
    output_base_path = "/kaggle/working/"

    # Use the Ollama solution verifier
    verifier = OllamaSolutionVerifier(model_name="gpt-oss:20b")
    processor = SolutionVerificationProcessor(verifier, failed_folder=os.path.join(output_base_path, "failed_verifications"))

    # # Process dataset
    # results, out_folder = processor.process_dataset(
    #     dataset_path,
    #     output_base_path,
    #     start_idx=0,
    #     end_idx=150  # Test with first 10
    # )
    indices_to_run = [0, 1, 23, 29, 90, 105]

    all_results = []
    out_folder = None  # placeholder
    
    for idx in indices_to_run:
        res, out_folder = processor.process_dataset(
            dataset_path,
            output_base_path,
            start_idx=idx,
            end_idx=idx+1  # only that index
        )
        all_results.extend(res)
    
    # Final results assignment (maintains your style)
    results, out_folder = all_results, out_folder
    
    print("Done. Results saved to:", out_folder)


if __name__ == "__main__":
    main()
```
