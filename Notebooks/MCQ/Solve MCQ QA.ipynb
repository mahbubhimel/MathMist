```python
import json
import os
import re
import time
from datetime import datetime
from fractions import Fraction
from tqdm import tqdm

# Ollama client (make sure Ollama daemon is running and 'ollama' Python package is installed)
try:
    import ollama
except Exception as e:
    raise ImportError("The 'ollama' package is required. Install it and make sure the Ollama daemon is running.") from e

# -------------------------
# Enhanced Answer Extractor with improved pattern matching
# -------------------------
class AnswerExtractor:
    @staticmethod
    def extract_answer_letter(content: str) -> str:
        """
        Enhanced function to extract an answer letter or numeric boxed answer from a model's output.
        It returns either a letter (A-D) or a numeric string (e.g., '14' or '1/3') if that's what was boxed.
        """
        if not content:
            return ""

        # Prefer explicit "Final Answer: X" patterns and take the last occurrence
        m = re.findall(r'Final Answer:\s*([A-Da-d])\b', content, re.IGNORECASE)
        if m:
            return m[-1].upper()

        # Common boxed LaTeX patterns for a LETTER
        latex_letter_patterns = [
            r"\\boxed\{([A-Da-d])\}",
            r"\$\\boxed\{([A-Da-d])\}\$",
            r"\\boxed\{\\text\{([A-Da-d])\}\}",
            r"\$\\boxed\{\\text\{([A-Da-d])\}\}\$",
        ]
        for p in latex_letter_patterns:
            mm = re.search(p, content)
            if mm:
                return mm.group(1).upper()

        # Look for "Answer: X" or similar single-letter statements (take last if multiple)
        m = re.findall(r'\bAnswer\s*[:\-]?\s*([A-Da-d])\b', content, re.IGNORECASE)
        if m:
            return m[-1].upper()

        # Sentence-based phrases
        m = re.findall(r'The (?:correct )?answer is\s*([A-Da-d])\b', content, re.IGNORECASE)
        if m:
            return m[-1].upper()

        # Boxed numeric answers like \boxed{14} or \boxed{1/3} -> return numeric as string
        mm = re.search(r"\\boxed\{([^\}]+)\}", content)
        if mm:
            boxed = mm.group(1).strip()
            # If the boxed content is a single letter, return it as letter
            if re.fullmatch(r'[A-Da-d]', boxed):
                return boxed.upper()
            return boxed  # could be numeric or expression; return so caller can try to map

        # If nothing matched, inspect the last ~10 lines for an isolated letter
        lines = content.strip().split('\n')
        for line in reversed(lines[-10:]):
            l = line.strip()
            if l in ['A', 'B', 'C', 'D']:
                return l
            # single letter occurrence on a line
            m = re.findall(r'\b([A-Da-d])\b', l)
            if len(m) == 1:
                return m[0].upper()

        # as last resort, find any occurrence and return last
        all_letters = re.findall(r'\b([A-Da-d])\b', content)
        if all_letters:
            return all_letters[-1].upper()

        return ""

    @staticmethod
    def extract_final_answer(text: str) -> str:
        """Extract the content inside a <final>...</final> tag or fall back to enhanced extraction."""
        if not text:
            return ""

        m = re.search(r'<final>(.*?)</final>', text, re.DOTALL | re.IGNORECASE)
        if m:
            return m.group(1).strip()

        return AnswerExtractor.extract_answer_letter(text)

    @staticmethod
    def try_parse_numeric(s: str):
        """
        Try to parse a numeric-like string into a Fraction or float.
        Returns a Fraction if possible, else float, else None.
        """
        if not s:
            return None
        s = s.strip()
        # strip common wrappers
        s = re.sub(r'^[\$\\()\s]+|[\$\\()\s]+$', '', s)
        # Try Fraction for rational strings like '1/3'
        try:
            if '/' in s:
                return Fraction(s)
        except Exception:
            pass
        # Try to extract numeric part
        try:
            # handle percentages
            if s.endswith('%'):
                return float(s[:-1]) / 100.0
            return float(s)
        except Exception:
            return None

    @staticmethod
    def map_numeric_to_letter(numeric_value: str, choices_mapping: dict) -> str:
        """
        Map a numeric value (string or numeric) to the corresponding letter choice.
        choices_mapping: dict of letter -> body (string)
        Returns letter or empty string.
        """
        if numeric_value is None:
            return ''
        num_parsed = AnswerExtractor.try_parse_numeric(str(numeric_value))
        if num_parsed is None:
            return ''

        # Compare against each choice
        for letter, body in choices_mapping.items():
            # clean body
            b = re.sub(r'\\boxed\{([^\}]+)\}', r'\1', str(body))
            b = b.strip()
            b_parsed = AnswerExtractor.try_parse_numeric(b)
            if b_parsed is None:
                # try to evaluate simple expressions (avoid eval for safety) - skip
                continue
            # Compare numeric types (Fraction compares to Fraction/float)
            try:
                # Convert to float for tolerance comparison if fractions are involved
                if isinstance(num_parsed, Fraction) or isinstance(b_parsed, Fraction):
                    # Compare as Fraction to avoid floating issues
                    if Fraction(num_parsed) == Fraction(b_parsed):
                        return letter
                else:
                    if abs(float(num_parsed) - float(b_parsed)) < 1e-6:
                        return letter
            except Exception:
                continue
        return ''


# -------------------------
# Ollama-based MCQ Math Solver with Enhanced Prompt
# -------------------------
class OllamaMCQMathSolver:
    def __init__(self, model_name: str = "gpt-oss:20b"):
        self.model_name = model_name
        self.client = ollama.Client()

    def _create_prompt(self, question: str, choices: list[str]) -> str:
        choices_block = "\n".join(choices)
        instructions = f"""You are an expert mathematician and MCQ solver in **Bangla/Bengali** language. Provide a concise answer to the problem below, then select the correct answer choice.

DO NOT provide chain-of-thought, step-by-step reasoning, or internal deliberation. If you include any justification, it must be a single short sentence preceded by 'Explanation:

Problem: {question}

Answer Choices:
{choices_block}

⚠️ CRITICAL OUTPUT FORMAT REQUIREMENT ⚠️
YOU MUST END YOUR RESPONSE WITH EXACTLY ONE OF THESE FOUR LINES:
Final Answer: A
Final Answer: B
Final Answer: C
Final Answer: D

❌ ABSOLUTELY FORBIDDEN - DO NOT USE:
- "The final answer is $\\boxed{{14}}$"
- "The final answer is $\\boxed{{A}}$"
- "$\\boxed{{A}}$"
- "\\boxed{{A}}"
- "(A)"
- "A is correct."
- "Final Answer: The answer is A"
- Any LaTeX formatting
- Any mathematical notation
- Any additional text after the letter

✅ REQUIRED FORMAT EXAMPLES:
If you determine the answer is choice A: "Final Answer: A"
If you determine the answer is choice B: "Final Answer: B"
If you determine the answer is choice C: "Final Answer: C"
If you determine the answer is choice D: "Final Answer: D"

IMPORTANT: Your response must end with exactly "Final Answer: [SINGLE LETTER]" - nothing else on that line. Do not include any boxed notation, LaTeX, or mathematical formatting in your final line.

Begin your concise answer now and remember to end with the exact required format."""
        return instructions

    def _generate_once(self, prompt_text: str, deterministic: bool = True, max_retries: int = 3, backoff: int = 2) -> str:
        final_prompt = prompt_text
        if deterministic:
            final_prompt = (
                "DETERMINISTIC MODE: Do not sample. Provide exactly the requested output.\n\n" +
                prompt_text
            )

        attempt = 0
        last_err = None
        while attempt < max_retries:
            try:
                # different ollama clients may have different signatures; try two forms
                try:
                    resp = self.client.chat(model=self.model_name,
                                            messages=[{"role": "user", "content": final_prompt}],
                                            think=False)
                except TypeError:
                    resp = self.client.chat(self.model_name,
                                            messages=[{"role": "user", "content": final_prompt}],
                                            think=False)
                # parse response
                if isinstance(resp, dict):
                    if 'message' in resp and isinstance(resp['message'], dict):
                        return (resp['message'].get('content', '') or '').strip()
                    if 'choices' in resp and resp['choices']:
                        choice = resp['choices'][0]
                        if isinstance(choice, dict) and 'message' in choice and isinstance(choice['message'], dict):
                            return (choice['message'].get('content', '') or '').strip()
                        return str(choice)
                    return str(resp)

                # obj-like
                try:
                    return (resp.message.content or '').strip()
                except Exception:
                    return str(resp)

            except Exception as e:
                last_err = e
                attempt += 1
                if attempt >= max_retries:
                    return f"ERROR_CALLING_OLLAMA: {e}"
                time.sleep(backoff ** attempt)
        return f"ERROR_CALLING_OLLAMA: {last_err}"

    def solve_mcq(self, question: str, choices: list[str]) -> dict:
        prompt = self._create_prompt(question, choices)
        output = self._generate_once(prompt, deterministic=True)
        extracted = AnswerExtractor.extract_final_answer(output)
        return {"raw_output": output, "extracted": extracted}


# -------------------------
# Helper utilities for MCQ evaluation
# -------------------------

def normalize_choice_token(token: str) -> str:
    """Normalize various formats to a single uppercase letter (A, B, C, ...)."""
    if not token:
        return ""
    token = str(token).strip()
    # direct patterns like "A", "A.", "(A)", "Final Answer: A"
    m = re.search(r'^.*?\(?\s*([A-Za-z])\s*\)?[\.)\:\-]*\s*$', token)
    if m:
        return m.group(1).upper()
    # fallback: find any standalone letter and choose the last one
    letters = re.findall(r'\b([A-Za-z])\b', token)
    return letters[-1].upper() if letters else ""


def match_choice_by_text(extracted_text: str, choices: list[str]) -> str:
    """Try to match the extracted text to one of the choice bodies."""
    if not extracted_text:
        return ""

    # build mapping letter -> body with enumeration fallback
    mapping = {}
    for i, ch in enumerate(choices):
        ch = ch.strip()
        m = re.match(r'^\s*([A-Za-z])\s*(?:[.)])\s*(.*)$', ch)
        if m:
            letter = m.group(1).upper()
            body = m.group(2).strip()
        else:
            letter = chr(ord('A') + i)
            body = ch
        mapping[letter] = body

    # Clean extracted_text for comparison
    ex = re.sub(r'[^0-9A-Za-z\\/\^_\.{}\-\s]', ' ', extracted_text).strip().lower()

    # try exact substring matches
    for letter, body in mapping.items():
        body_clean = re.sub(r'[^0-9A-Za-z\\/\^_\.{}\-\s]', ' ', body).strip().lower()
        if body_clean and (body_clean in ex or ex in body_clean):
            return letter

    # token-level similarity
    ex_words = set([w for w in ex.split() if len(w) > 1])
    best = (None, 0.0)
    for letter, body in mapping.items():
        b_words = set([w for w in re.sub(r'[^0-9A-Za-z\\/\^_\.{}\-\s]', ' ', body).lower().split() if len(w) > 1])
        if not b_words:
            continue
        common = len(ex_words & b_words)
        score = common / max(1, len(b_words))
        if score > best[1]:
            best = (letter, score)

    if best[0] and best[1] >= 0.5:
        return best[0]

    return ""


# -------------------------
# Dataset Processor adapted for MCQ-only
# -------------------------
class DatasetProcessorMCQ:
    def __init__(self, solver: OllamaMCQMathSolver, failed_folder=None):
        self.solver = solver
        self.failed_folder = failed_folder or "failed_extractions_mcq"
        os.makedirs(self.failed_folder, exist_ok=True)
        # single consolidated failed JSONL file to avoid many tiny files
        self.failed_file_path = os.path.join(self.failed_folder, 'failed_mcq.jsonl')

    def _load_dataset_stream(self, dataset_path: str):
        with open(dataset_path, 'r', encoding='utf-8') as f:
            for line in f:
                if line.strip():
                    yield json.loads(line)

    def process_dataset(self, dataset_path: str, output_base_path: str, start_idx=0, end_idx=None,
                        folder_name=None, create_timestamped_folder=True):
        """
        Stream the dataset and process entries with a comprehensive tqdm progress bar.
        This implementation first attempts to count total lines (so tqdm can show a total). If counting
        fails (e.g., due to non-file stream), tqdm will run without a total but still show rate/progress.
        """
        # Try to determine total number of non-empty lines for a meaningful progress bar
        total = None
        try:
            with open(dataset_path, 'r', encoding='utf-8') as f:
                total = sum(1 for line in f if line.strip())
        except Exception:
            total = None

        stream = self._load_dataset_stream(dataset_path)

        # Prepare output folder
        output_folder = self._create_output_folder(output_base_path, folder_name, start_idx,
                                                   end_idx or 'end', create_timestamped_folder)
        results = []

        print(f"Processing MCQ problems starting at {start_idx}")
        print(f"Output will be saved in: {output_folder}")

        # Open failed file for append (consolidated failures)
        failed_fh = open(self.failed_file_path, 'a', encoding='utf-8')

        processed = 0
        correct_count = 0

        # If total is known, adjust for start_idx/end_idx bounds to show accurate total
        display_total = None
        if total is not None:
            start = max(0, start_idx)
            end = total if end_idx is None else min(total, end_idx)
            display_total = max(0, end - start)

        desc = "Processing MCQs"
        with tqdm(total=display_total, desc=desc, unit="q") as pbar:
            try:
                for idx, problem in enumerate(stream):
                    # Skip until start_idx (but still advance progress if we know total)
                    if idx < start_idx:
                        if display_total is not None:
                            pbar.update(1)
                        continue
                    if end_idx is not None and idx >= end_idx:
                        break

                    try:
                        entry = self._process_single(idx, problem)
                        results.append(entry)
                        if entry.get('is_correct'):
                            correct_count += 1
                    except Exception as e:
                        print(f"Error at index {idx}: {e}")
                        entry = self._create_error_entry(idx, problem, str(e))
                        results.append(entry)

                    processed += 1
                    pbar.update(1)

                    # update postfix with lightweight stats
                    if processed > 0:
                        acc = (correct_count / processed * 100)
                        pbar.set_postfix({'processed': processed, 'correct': correct_count, 'acc': f"{acc:.2f}%"})

                    if processed % 50 == 0:
                        self._save_intermediate_results(results, output_folder, processed)

                final_path = self._save_final_results(results, output_folder, start_idx, end_idx or idx + 1)
                self._create_summary_file(results, output_folder, dataset_path, start_idx, end_idx or idx + 1)
                return results, output_folder

            finally:
                failed_fh.close()

    def _process_single(self, idx: int, problem: dict) -> dict:
        question = problem.get('Question', '')
        choices = problem.get('Multiple Choices', []) or []
        correct_answer = problem.get('Correct Answer', '')
        answer_type = problem.get('Answer Type', '')
        eval_method = problem.get('Evaluation Method', '')

        # Build mapping letter -> body with enumeration fallback (used for numeric matching)
        choices_mapping = {}
        for i, ch in enumerate(choices):
            ch = ch.strip()
            m = re.match(r'^\s*([A-Za-z])\s*(?:[.)])\s*(.*)$', ch)
            if m:
                letter = m.group(1).upper()
                body = m.group(2).strip()
            else:
                letter = chr(ord('A') + i)
                body = ch
            choices_mapping[letter] = body

        # Call solver
        solver_res = self.solver.solve_mcq(question, choices)
        raw_output = solver_res.get('raw_output', '')
        extracted = solver_res.get('extracted', '')

        # try normalization
        letter = normalize_choice_token(extracted)

        # If extracted is not a letter but looks numeric, try numeric->letter mapping
        if not letter and extracted:
            # if extracted looks numeric or fraction
            if re.search(r'^[0-9\./\s%-]+$', extracted):
                mapped = AnswerExtractor.map_numeric_to_letter(extracted, choices_mapping)
                if mapped:
                    letter = mapped

        if not letter:
            # try to match by text
            letter = match_choice_by_text(extracted, choices)

        # as last resort, try to extract letter from the raw output
        if not letter:
            letter = normalize_choice_token(raw_output)
        if not letter:
            letter = match_choice_by_text(raw_output, choices)

        # normalize correct_answer to a single letter
        correct_letter = normalize_choice_token(str(correct_answer))

        # If dataset's correct answer is full text, try to map it to a letter
        if not correct_letter and correct_answer:
            corr_by_text = match_choice_by_text(str(correct_answer), choices)
            if corr_by_text:
                correct_letter = corr_by_text

        is_correct = False
        if letter and correct_letter:
            is_correct = (letter == correct_letter)

        # Save failed examples for inspection into a consolidated JSONL file
        if not letter:
            # append to single JSONL failed file
            fname = os.path.join(self.failed_folder, 'failed_mcq.jsonl')
            with open(fname, 'a', encoding='utf-8') as f:
                json.dump({
                    'index': idx,
                    'question': question,
                    'choices': choices,
                    'raw_output': raw_output,
                    'extracted': extracted,
                    'correct_answer': correct_answer,
                    'answer_type': answer_type,
                    'eval_method': eval_method,
                    'timestamp': datetime.now().isoformat()
                }, f, ensure_ascii=False)
                f.write('\n')

        entry = {
            'problem_index': idx,
            'question': question,
            'choices': choices,
            'raw_output': raw_output,
            'extracted_raw': extracted,
            'predicted_letter': letter,
            'correct_letter': correct_letter,
            'is_correct': is_correct,
            'answer_type': answer_type,
            'eval_method': eval_method,
        }

        return entry

    def _create_output_folder(self, base_path, folder_name, start_idx, end_idx, add_timestamp):
        if folder_name is None:
            folder_name = f"mcq_results_{start_idx}_to_{end_idx}"
        if add_timestamp:
            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
            folder_name = f"{folder_name}_{timestamp}"
        output_folder = os.path.join(base_path, folder_name)
        os.makedirs(output_folder, exist_ok=True)
        return output_folder

    def _save_intermediate_results(self, results, output_folder, count):
        p = os.path.join(output_folder, f'intermediate_mcq_{count}.json')
        with open(p, 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=2)
        print(f"Saved intermediate results to {p}")

    def _save_final_results(self, results, output_folder, start_idx, end_idx):
        p = os.path.join(output_folder, f'final_mcq_{start_idx}_to_{end_idx-1}.json')
        with open(p, 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=2)
        print(f"Saved final results to {p}")
        return p

    def _create_summary_file(self, results, output_folder, dataset_path, start_idx, end_idx):
        total = len(results)
        correct = len([r for r in results if r.get('is_correct')])
        acc = (correct / total * 100) if total else 0.0

        summary = {
            'dataset_path': dataset_path,
            'start_index': start_idx,
            'end_index': end_idx - 1,
            'total_processed': total,
            'correct_count': correct,
            'accuracy_percent': f"{acc:.2f}%",
            'processing_timestamp': datetime.now().isoformat()
        }

        p = os.path.join(output_folder, 'summary_mcq.json')
        with open(p, 'w', encoding='utf-8') as f:
            json.dump(summary, f, ensure_ascii=False, indent=2)
        print(f"Saved summary to {p} | Accuracy: {summary['accuracy_percent']}")

    def _create_error_entry(self, idx, problem, error_msg):
        return {
            'problem_index': idx,
            'question': problem.get('Question', ''),
            'choices': problem.get('Multiple Choices', []),
            'raw_output': f"ERROR: {error_msg}",
            'extracted_raw': '',
            'predicted_letter': '',
            'correct_letter': normalize_choice_token(str(problem.get('Correct Answer', ''))),
            'is_correct': False,
            'answer_type': answer_type,
            'eval_method': eval_method
        }


# -------------------------
# Main example usage
# -------------------------

def main():
    dataset_path = '/kaggle/input/nctb-dataset/MCQ_Numerical_Symbolic_Bangla.jsonl'  # update to your MCQ JSONL
    output_base_path = '/kaggle/working/'

    solver = OllamaMCQMathSolver(model_name='gpt-oss:20b')
    processor = DatasetProcessorMCQ(solver,
                                   failed_folder=os.path.join(output_base_path, 'failed_extractions_mcq'))

    results, out_folder = processor.process_dataset(dataset_path, output_base_path,
                                                   start_idx=0, end_idx=1133)
    print('Done. Results saved to:', out_folder)


if __name__ == '__main__':
    main()
```
